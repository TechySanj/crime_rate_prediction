```python
  x=data[['land_area','percent_city','percent_senior','physicians','hospital_beds','graduates','work_force','income','region']]
y=data['crime_rate']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

Creating a regression model with the help of :

1. Linear Regression

2. Decision Tree Regression

3. Random Forest Regression

```python
#creating the model with Linear Regression
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
linear_predictions=regression.predict(x_test)
```

```python
regression.score(x,y)
```

```python
#creating the model with Decision_Tree_Regressor
from sklearn.tree import DecisionTreeRegressor
tree_regressor=DecisionTreeRegressor()
tree_regressor.fit(x_train,y_train)
```

```python
tree_predictions=tree_regressor.predict(x_test)
```

```python
tree_regressor.score(x,y)
```

```python
#creating the model with Random_Forest_Regressor
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
rfr_predictions=rfr.predict(x_test)
```

```python
rfr.score(x,y)
```
